```python
import numpy as np

# Importing standard Qiskit libraries
from qiskit import QuantumCircuit, transpile, Aer, IBMQ, QuantumRegister, ClassicalRegister, BasicAer, qiskit
from qiskit.tools.jupyter import *
from qiskit.visualization import *
from ibm_quantum_widgets import *
from qiskit.providers.aer import QasmSimulator

# Loading your IBM Quantum account(s)
provider = IBMQ.load_account()
```

```python
from numpy import pi
```

```python
qreg_q=QuantumRegister(4,'q')
creg_c=ClassicalRegister(4,'c')
circuit=QuantumCircuit(qreg_q, creg_c)

circuit.h(qreg_q[0])
circuit.h(qreg_q[1])
circuit.h(qreg_q[2])
circuit.h(qreg_q[3])

circuit.s(qreg_q[0])
circuit.t(qreg_q[1])
circuit.p(pi/8,qreg_q[2])
circuit.z(qreg_q[3])
```

```python
circuit.draw()
```

```python
from qiskit.quantum_info import Statevector
```

```python
sv=Statevector.from_label('0000')
sv_ev=sv.evolve(circuit)
sv_ev.draw('qsphere')
```

```python
backend=BasicAer.get_backend('statevector_simulator')
job=qiskit.execute(circuit,backend,shots=1024)
result=job.result()
sv_ev2=result.get_statevector(circuit)
plot_state_qsphere(sv_ev2)
```
