```python
import numpy as np

# Importing standard Qiskit libraries
from qiskit import QuantumCircuit, transpile, Aer, IBMQ, qiskit
from qiskit.tools.jupyter import *
from qiskit.visualization import *
from ibm_quantum_widgets import *
from qiskit.providers.aer import QasmSimulator

# Loading your IBM Quantum account(s)
provider = IBMQ.load_account()
```

```python
qc=QuantumCircuit(1)
qc.x(0)
sv_sim=Aer.get_backend('unitary_simulator')
job=qiskit.execute(qc,sv_sim)
output_vector=job.result().get_unitary(qc,decimals=3)
print(output_vector)
```

```python
from math import sqrt
qc2=QuantumCircuit(2)
state_vector=[1/sqrt(2),0,0,1/sqrt(2)]
qc2.initialize(state_vector,[0,1])

sv_sim=Aer.get_backend('unitary_simulator')
job=qiskit.execute(qc2,sv_sim)
output_vector=job.result().get_unitary(qc2,decimals=3)
print(output_vector)

plot_state_city(output_vector, color=['pink','green'])
```
